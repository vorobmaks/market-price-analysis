**Завдання 1. Розширений статистичний аналіз даних.**

На першому етапі було виконано фільтрацію набору даних за умовою:

mp_year ≤ 2000

Після застосування даного фільтра отримано 14 577 записів, що використовувались у подальшому аналізі. Для кожного товару з поля cm_name було обчислено базові статистичні характеристики показника mp_price.

У вихідному наборі даних ціни подані в різних національних валютах. Тому для забезпечення коректності статистичного аналізу всі значення цін було приведено до єдиної валюти — USD.

Отримані результати показали, що найвищі ціни зафіксовані для товарів, пов’язаних із тваринництвом та оплатою праці. Зокрема, товар Livestock (sheep, one-year-old alive female) - Retail має середню ціну близько 12 326 USD, а максимальна вартість перевищує 24 850 USD. Також високі значення мають Meat (camel) - Retail із середньою ціною приблизно 1 181 USD, Wage (non-qualified labour, non-agricultural) - Retail — близько 995 USD та Livestock (Goat) - Retail — близько 199 USD.

Водночас основні продовольчі товари мають значно нижчі значення цін. Наприклад, Wheat - Retail характеризується середнім рівнем ціни близько 0.10 USD, а більшість спостережень зосереджена в інтервалі 0.08–0.11 USD, що свідчить про стабільність цін. Аналогічну динаміку демонструють Rice - Retail, Maize - Retail та Millet - Retail, де середні ціни не перевищують 0.3 USD.

Boxplot підтвердив, що ринки основних харчових продуктів є стабільними, тоді як Fuel (diesel) - Retail та Rice (imported) - Retail мають значні коливання цін і численні викиди. Крім того, перехресна таблиця показала чіткий поділ товарів між сегментами роздрібного та оптового продажу, що вказує на різну структуру ринку для окремих категорій товарів.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("global_food_prices (1).csv", low_memory=False)
df_filt = df[df['mp_year'] <= 2000].copy()
print("Кількість записів після фільтрації:", len(df_filt))

rates = {
    "XOF": 600,
    "INR": 80,
    "SOS": 570,
    "MZN": 64,
    "PHP": 55,
    "AFN": 86,
    "PEN": 3.8,
    "MXN": 18,
    "COP": 4000,
    "AMD": 400,
    "ETB": 56,
    "NIO": 36,
    "BDT": 110,
    "GTQ": 7.5,
    "ZAR": 18,
    "THB": 36,
    "RWF": 1300
}

df_filt["rate"] = df_filt["cur_name"].map(rates)

df_filt["mp_price_usd"] = df_filt["mp_price"] / df_filt["rate"]

stats = df_filt.groupby('cm_name')['mp_price_usd'].agg(
    MIN='min',
    MAX='max',
    MEAN='mean',
    Q1=lambda x: x.quantile(0.25),
    MEDIAN=lambda x: x.quantile(0.5),
    Q3=lambda x: x.quantile(0.75),
    VARIANCE='var'
)

print("\n=== Статистика по товарах (USD) ===")
display(stats)

crosstab = pd.crosstab(df_filt['cm_name'], df_filt['pt_name'])
print("\n=== Crosstab cm_name × pt_name ===")
display(crosstab)

top_cm = df_filt['cm_name'].value_counts().head(10).index
subset = df_filt[df_filt['cm_name'].isin(top_cm)]

plt.figure(figsize=(12, 6))
subset.boxplot(column='mp_price_usd', by='cm_name', rot=90)
plt.title("Boxplot mp_price_usd (mp_year ≤ 2000, ТОП-10 товарів)")
plt.suptitle("")
plt.ylabel("Ціна (USD)")
plt.xlabel("Товари")
plt.grid()
plt.show()

PRODUCT = "Wheat - Retail"
product_data = df_filt[df_filt['cm_name'] == PRODUCT]

print("\n=== Статистика для товару (USD):", PRODUCT, "===")
display(product_data['mp_price_usd'].describe())

plt.figure(figsize=(5,6))
plt.boxplot(product_data['mp_price_usd'])
plt.title(f"Boxplot для {PRODUCT} (USD)")
plt.ylabel("Ціна (USD)")
plt.grid()
plt.show()

df_filt.to_csv("global_food_prices_usd.csv", index=False)

print("Файл збережено як: global_food_prices_usd.csv")
```

**Завдання 3.Обробка пропусків та кодування змінних**

Найбільша кількість пропущених значень виявлена у полі mp_commoditysource (100%), а також у полі adm1_name (близько 30%). Оскільки колонка mp_commoditysource повністю складається з пропусків, її значення не несли інформаційної цінності для аналізу та не впливали на результати, а пропуски в полі adm1_name були заповнені найбільш імовірним значенням (модою).

Текстову змінну adm0_name було перекодовано в числовий формат шляхом створення нової колонки adm0_name_code, де кожній країні присвоєно унікальний числовий код від 1 до 19. Побудований scatter plot для mp_year та adm0_name_code демонструє розподіл спостережень по країнах у часі та показує, що дані присутні для всіх закодованих країн у вибраному періоді.

Таким чином, пропуски в даних було успішно оброблено, а категоріальна змінна приведена до числового формату.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_filt = df[df["mp_year"] <= 2000].copy()

missing_count = df_filt.isna().sum()
missing_percent = (df_filt.isna().mean() * 100).round(2)

missing_table = pd.DataFrame({
    "Кількість помилок": missing_count,
    "Відсоток": missing_percent
})

print("\n=== ПРОПУСКИ В ДАНИХ ===")
display(missing_table[missing_table["Missing count"] > 0])

cat_cols = df_filt.select_dtypes(include="object").columns
for col in cat_cols:
    df_filt[col] = df_filt[col].fillna(df_filt[col].mode()[0])
print("\nПропуски оброблено.")

ENCODE_COL = "adm0_name"

df_filt[ENCODE_COL] = df_filt[ENCODE_COL].astype(str).str.strip()

codes, uniques = pd.factorize(df_filt[ENCODE_COL])
df_filt[ENCODE_COL + "_code"] = codes + 1

encoder_map = pd.DataFrame({
    "code": range(1, len(uniques) + 1),
    ENCODE_COL: uniques
})

print("\n=== ТАБЛИЦЯ КОДУВАННЯ ===")
display(encoder_map.head(20))

plt.figure(figsize=(8,5))
plt.scatter(df_filt["mp_year"], df_filt[ENCODE_COL + "_code"])
plt.title(f"Scatter plot: mp_year vs {ENCODE_COL}_code (mp_year <= 2000)")
plt.xlabel("Рік")
plt.ylabel(f"{ENCODE_COL}_code")
plt.grid()
plt.show()
```

                   mp_price=−58654.24+29.40⋅mp_year

Це означає, що за оцінкою моделі середня ціна Wheat зростає приблизно на 29.4 одиниці на рік.

**Завдання 9. Аналіз чутливості моделі**

Для оцінки чутливості моделі було змінено значення mp_year на ±1, ±2, ±3 роки відносно базового року та зафіксовано зміну прогнозованого mp_price. Аналіз показав, що модель реагує лінійно:

збільшення року → пропорційне зростання ціни,

зменшення року → симетричне падіння прогнозу.

Побудовані графіки демонструють майже лінійну залежність між зміною року та відносною зміною ціни, що відповідає природі лінійної регресії.

**Завдання 11. Розширений аналіз чутливості**

Розширений аналіз чутливості було проведено з метою кількісної оцінки впливу змінної mp_year на результат прогнозування mp_price. Аналіз виконано шляхом зміщення значення року на ±1, ±2 та ±3 роки відносно базового значення з подальшим перерахунком прогнозованої ціни.

Отримані результати показали, що модель є суттєво чутливою до зміни змінної mp_year. Зменшення року на один рік призводить до падіння прогнозованої ціни приблизно на 21%, на два роки — на 41%, а на три роки — майже на 62%. Аналогічно, зміщення року у майбутнє дає симетричне зростання прогнозу: +1 рік — близько +21%, +2 роки — +41%, +3 роки — майже +62%.

Таким чином встановлено, що змінна mp_year має найбільший вплив на прогнозоване значення mp_price.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

PRODUCT = "Wheat"
df_filt = df[(df["mp_year"] <= 2000) &
             (df["cm_name"].str.contains(PRODUCT, case=False, na=False))].copy()

print("Кількість записів після фільтрації:", len(df_filt))

df_filt = df_filt.dropna(subset=["mp_year", "mp_price"])

yearly = (
    df_filt
    .groupby("mp_year", as_index=False)["mp_price"]
    .mean()
    .sort_values("mp_year")
)

print("\nПерші роки після агрегації:")
display(yearly.head())

X = yearly[["mp_year"]].values
y = yearly["mp_price"].values

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

MAE = mean_absolute_error(y, y_pred)
RMSE = np.sqrt(mean_squared_error(y, y_pred))
R2 = r2_score(y, y_pred)

print("\n=== МЕТРИКИ МОДЕЛІ ===")
print(f"MAE  = {MAE:.3f}")
print(f"RMSE = {RMSE:.3f}")
print(f"R^2  = {R2:.4f}")

print("\nРівняння моделі:")
print(f"mp_price = {model.intercept_:.4f} + {model.coef_[0]:.4f} * mp_year")

last_year = int(yearly["mp_year"].max())
future_years = np.array([[last_year + 1],
                          [last_year + 2],
                          [last_year + 3]])

future_pred = model.predict(future_years)

forecast = pd.DataFrame({
    "mp_year": future_years.flatten(),
    "predicted_price": future_pred
})

print("\n=== ПРОГНОЗ НА 3 РОКИ ВПЕРЕД ===")
display(forecast)

plt.figure(figsize=(10,5))
plt.scatter(yearly["mp_year"], y, label="Фактичні дані")
plt.plot(yearly["mp_year"], y_pred, label="Лінійна регресія")
plt.plot(forecast["mp_year"], forecast["predicted_price"], "ro--", label="Прогноз")
plt.title(f"Linear Regression for {PRODUCT}")
plt.xlabel("mp_year")
plt.ylabel("mp_price")
plt.grid()
plt.legend()
plt.show()

def my_sens_lin_year(model, base_year, shifts):
    base_year = float(base_year)
    base_pred = model.predict([[base_year]])[0]
    rows = []

    for d in shifts:
        new_year = base_year + d
        pred = model.predict([[new_year]])[0]

        rows.append({
            "Δ_year": d,
            "mp_year": round(new_year, 2),
            "predicted_price": pred,
            "Δ_price": pred - base_pred,
            "%_change": (pred - base_pred) / base_pred * 100
        })

    return pd.DataFrame(rows)

base_year = yearly["mp_year"].mean()
year_shifts = [-3, -2, -1, 1, 2, 3]

sens_year = my_sens_lin_year(model, base_year, year_shifts)

print("\n=== АНАЛІЗ ЧУТЛИВОСТІ (ЗА РОКАМИ) ===")
display(sens_year)

plt.figure(figsize=(7,5))
plt.plot(sens_year["Δ_year"], sens_year["%_change"], marker="o")
plt.axhline(0, color="gray", linestyle="--")
plt.xlabel("Δ mp_year (роки)")
plt.ylabel("Зміна mp_price (%)")
plt.title("Sensitivity to mp_year (line plot)")
plt.grid()
plt.show()

plt.figure(figsize=(7,5))
plt.bar(sens_year["Δ_year"], sens_year["%_change"])
plt.axhline(0, color="gray", linestyle="--")
plt.xlabel("Δ mp_year (роки)")
plt.ylabel("Зміна mp_price (%)")
plt.title("Sensitivity to mp_year (bar plot)")
plt.grid(axis="y")
plt.show()

def my_sens_year_extended(model, base_year, deltas):
    base_pred = model.predict([[base_year]])[0]
    rows = []

    for d in deltas:
        new_year = base_year + d
        pred = model.predict([[new_year]])[0]

        rows.append({
            "Δ_year": d,
            "mp_year": int(new_year),
            "predicted_price": pred,
            "Δ_price": pred - base_pred,
            "%_change": (pred - base_pred) / base_pred * 100
        })

    return pd.DataFrame(rows)

base_year = yearly["mp_year"].max()
year_deltas = [-3, -2, -1, 1, 2, 3]

sens_ext = my_sens_year_extended(model, base_year, year_deltas)

print("\n=== РОЗШИРЕНИЙ АНАЛІЗ ЧУТЛИВОСТІ ===")
display(sens_ext)

plt.figure(figsize=(8,5))
plt.bar(sens_ext["Δ_year"].astype(str), sens_ext["%_change"])
plt.axhline(0, color="black", linewidth=0.8)
plt.xlabel("Зміна mp_year (роки)")
plt.ylabel("Зміна прогнозованої mp_price, %")
plt.title(f"Sensitivity analysis for {PRODUCT} (by year shift)")
plt.grid(axis="y", linestyle="--")
plt.show()
```

**Завдання 11. Розширений аналіз чутливості**

Чутливість до параметра mp_year

Результати аналізу чутливості за роками демонструють виражену монотонну тенденцію. Зі зменшенням значення mp_year прогнозована ціна систематично зменшується, а зі збільшенням — зростає. Так, при зміщенні року назад на три роки модель прогнозує зниження ціни приблизно на 40 одиниць, тоді як зсув вперед на три роки призводить до зростання ціни майже на 44 одиниці.

Отримана залежність свідчить, що змінна mp_year є домінуючою ознакою в моделі, і саме вона формує основний довгостроковий тренд динаміки цін. Така поведінка моделі узгоджується з історичними даними та вказує на наявність зростаючої тенденції в ціні для досліджуваного регіону.

Чутливість до параметра mp_month

Аналіз впливу місяця показує, що місячні зсуви мають суттєво менший ефект порівняно з роками. Зміна місяця в межах одного року призводить лише до незначних коливань у прогнозованій ціні, які носять сезонний характер.

При зсуві на попередні місяці модель прогнозує незначне підвищення ціни, однак для позитивних зсувів у межах поточного року спостерігаються як мінімальні зростання, так і слабке зниження ціни. Водночас перехід на наступний рік при зсуві +6 місяців викликає більш різкий стрибок ціни, що ще раз підтверджує визначальний вплив саме року, а не місяця.

Отже, змінна mp_month відіграє допоміжну роль та відображає скоріше сезонні коливання, ніж довгострокову тенденцію.

Комбінована чутливість (mp_year + mp_month)

Комбінований аналіз зсувів року і місяця показав кумулятивний ефект. При одночасному збільшенні року та місяця модель демонструє значне зростання прогнозованої ціни. Наприклад, зсув на +3 роки та +9 місяців призводить до підвищення ціни більш ніж на 60 одиниць, що відповідає приросту понад 30%.

Натомість зворотні сценарії, де рік і місяць зменшуються, дають протилежний ефект. При переході на три роки і дев’ять місяців у минуле ціна зменшується майже на 55 одиниць, що відповідає втраті майже третини початкової вартості.

Це підтверджує, що чинники часу мають мультиплікативний ефект у моделі, де вплив року значно перевищує вплив місяця.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

REGION = "Tillaberi"

df_filt = df[(df["mp_year"] <= 2000) &(df["adm1_name"] == REGION)].copy()
print("Кількість записів:", len(df_filt))

df_filt = df_filt.dropna(subset=["mp_year","mp_month","mp_price"])

gb = (
    df_filt
    .groupby(["mp_year","mp_month"], as_index=False)["mp_price"]
    .mean()
    .sort_values(["mp_year","mp_month"])
)

X = gb[["mp_year","mp_month"]].values
y = gb["mp_price"].values

lr = LinearRegression()
mlp = Pipeline([
    ("scaler", StandardScaler()),
    ("mlp", MLPRegressor(hidden_layer_sizes=(50,30),
                         max_iter=700,
                         random_state=42))
])

lr.fit(X, y)
mlp.fit(X, y)

pred_lr = lr.predict(X)
pred_mlp = mlp.predict(X)

def metrics(y, p):
    return {
        "MAE": mean_absolute_error(y, p),
        "RMSE": np.sqrt(mean_squared_error(y, p)),
        "R²": r2_score(y, p)
    }

print("\n=== МЕТРИКИ ===")
print("Лінійна регресія:", metrics(y, pred_lr))
print("MLPRegressor:", metrics(y, pred_mlp))

t = gb["mp_year"] + (gb["mp_month"]-1)/12

plt.figure(figsize=(10,5))
plt.plot(t, y, "o-", label="Факт")
plt.plot(t, pred_lr, label="LinearRegression")
plt.plot(t, pred_mlp, "--", label="MLPRegressor")
plt.xlabel("Рік")
plt.ylabel("mp_price")
plt.title(f"Прогнозування для регіону {REGION}")
plt.grid()
plt.legend()
plt.show()

last_year = gb["mp_year"].max()

future = []
for i in range(1,37):
    future.append([
        last_year + (i-1)//12 + 1,
        (i-1)%12 + 1
    ])

future = np.array(future)

future_lr = lr.predict(future)
future_mlp = mlp.predict(future)

forecast = pd.DataFrame({
    "mp_year": future[:,0],
    "mp_month": future[:,1],
    "LR_pred": future_lr,
    "MLP_pred": future_mlp
})

print("\n=== ПРОГНОЗ НА 3 РОКИ ===")
display(forecast)
def sens_year_absolute(model, base, shifts):
    base = np.array(base, dtype=float)
    base_pred = model.predict([base])[0]
    rows = []

    for d in shifts:
        x = base.copy()
        x[0] += d   # зміна року

        pred = model.predict([x])[0]

        rows.append({
            "Δ_year": d,
            "mp_year": int(x[0]),
            "predicted": pred,
            "Δ_price": pred - base_pred
        })

    return pd.DataFrame(rows)


YEAR_SHIFTS = [-3, -2, -1, 1, 2, 3]
sens_year = sens_year_absolute(mlp, BASE, YEAR_SHIFTS)

print("\n=== ЧУТЛИВІСТЬ ДО mp_year (± роки) ===")
display(sens_year)

def sens_month_absolute(model, base, shifts):
    base = np.array(base, dtype=float)
    base_pred = model.predict([base])[0]

    rows = []

    for d in shifts:
        x = base.copy()

        # переводимо дату у абсолютні місяці
        total_months = int(x[0]) * 12 + (int(x[1]) - 1)

        # зсув у місяцях
        total_months += d

        # назад у (year, month)
        x[0] = total_months // 12
        x[1] = total_months % 12 + 1

        pred = model.predict([x])[0]

        rows.append({
            "Δ_month": d,
            "mp_year": int(x[0]),
            "mp_month": int(x[1]),
            "predicted": pred,
            "Δ_price": pred - base_pred
        })

    return pd.DataFrame(rows)

def sens_combined_abs(model, base, deltas):
    base = np.array(base, dtype=float)
    base_pred = model.predict([base])[0]
    rows = []

    for dy, dm in deltas:
        new = base.copy()

        # зсув року
        new[0] = new[0] + dy

        # зсув місяця
        new[1] = new[1] + dm

        # корекція місяців, якщо вихід за 1..12
        if new[1] > 12:
            new[0] += (new[1] - 1) // 12
            new[1] = ((new[1] - 1) % 12) + 1
        elif new[1] < 1:
            shift = abs(new[1]) // 12 + 1
            new[0] -= shift
            new[1] = 12 - (abs(new[1]) % 12)

        pred = model.predict([new])[0]

        rows.append({
            "Δyear": dy,
            "Δmonth": dm,
            "new_year": int(new[0]),
            "new_month": int(new[1]),
            "predicted": round(pred, 2),
            "Δmp_price": round(pred - base_pred, 2),
            "%change": round((pred - base_pred)/base_pred * 100, 2)
        })

    return pd.DataFrame(rows)

SCENARIOS = [
    (1, 3),
    (2, 6),
    (3, 9),
    (-1, -3),
    (-2, -6),
    (-3, -9)
]

sens_combo = sens_combined_abs(mlp, BASE, deltas=SCENARIOS)

print("\n=== КОМБІНОВАНА ЧУТЛИВІСТЬ (Δyear, Δmonth) ===")
display(sens_combo)

plt.figure(figsize=(7,5))
plt.bar(sens_year["Δ_year"], sens_year["Δ_price"])
plt.axhline(0, color="black")
plt.xlabel("Зміщення mp_year (роки)")
plt.ylabel("Зміна mp_price")
plt.title("Чутливість до mp_year (абсолютні зміни)")
plt.grid(axis="y")
plt.show()

plt.figure(figsize=(8,4))
labels = [f"{dy:+}y {dm:+}m" for dy, dm in zip(sens_combo["Δyear"], sens_combo["Δmonth"])]
plt.bar(labels, sens_combo["Δmp_price"])
plt.axhline(0, color="black")
plt.title("Комбінована чутливість моделі (Δyear + Δmonth)")
plt.ylabel("Зміна mp_price")
plt.xlabel("Сценарії зсувів")
plt.grid(axis="y", linestyle="--")

plt.show()
```

          mp_price(Kabul) = 2.37 + 0.95 · mp_price(Fayzabad)

**Розширений аналіз чутливості**

Розширений аналіз чутливості показав, що модель є симетрично чутливою до змін вхідної змінної. Зменшення ціни в Fayzabad на 20% призводить до падіння прогнозованої ціни в Kabul приблизно на 17.7%, а підвищення — до аналогічного зростання. Менші відхилення також демонструють лінійну залежність: ±10% зміни на ринку A спричиняє приблизно ±8.9% зміни на ринку B.

Отримані результати підтверджують тісний економічний зв’язок між дослідженими ринками.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

MARKET_A = "Fayzabad"
MARKET_B = "Kabul"
PRODUCT  = "Wheat"

df_filt = df[
    (df["mkt_name"].isin([MARKET_A, MARKET_B])) &
    (df["cm_name"].str.contains(PRODUCT, na=False))
].copy()

print("Кількість записів:", len(df_filt))

mean_prices = (
    df_filt
    .groupby(["mkt_name","cm_name"], as_index=False)["mp_price"]
    .mean()
)

print("\n=== СЕРЕДНІ ЦІНИ ПО РИНКАХ І ТОВАРАХ ===")
display(mean_prices.head())

grouped = (
    df_filt
    .groupby(["mp_year","mp_month","mkt_name"], as_index=False)["mp_price"]
    .mean()
)

pivot = grouped.pivot_table(
    index=["mp_year","mp_month"],
    columns="mkt_name",
    values="mp_price"
).dropna()

print("\n=== PIVOT_TABLE (Market A vs Market B) ===")
display(pivot.head())

X = pivot[[MARKET_A]].values
y = pivot[MARKET_B].values

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("\n=== ЛІНІЙНА МОДЕЛЬ ===")
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print("R²:", r2_score(y, y_pred))

t = np.arange(len(pivot))

plt.figure(figsize=(10,4))
plt.plot(t, pivot[MARKET_A], label=MARKET_A)
plt.plot(t, pivot[MARKET_B], label=MARKET_B)
plt.xlabel("Час")
plt.ylabel("Ціна")
plt.title(f"Тренди цін: {MARKET_A} vs {MARKET_B}")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(6,5))
plt.scatter(X, y, alpha=0.7)
plt.plot(X, y_pred, color="red")
plt.xlabel(f"Ціна на ринку {MARKET_A}")
plt.ylabel(f"Ціна на ринку {MARKET_B}")
plt.title("Регресія: A → B")
plt.grid()
plt.show()

def my_sens(model, base, percs):
    base = float(base)
    base_pred = model.predict([[base]])[0]
    rows = []

    for p in percs:
        new_x = base * (1 + p)
        pred = model.predict([[new_x]])[0]
        rows.append({
            "Δ_input_%": p*100,
            "new_price_A": new_x,
            "predicted_B": pred,
            "Δ_output_%": (pred - base_pred)/base_pred * 100
        })

    return pd.DataFrame(rows)

base_price = pivot[MARKET_A].mean()

percents = [-0.20, -0.10, -0.05, 0.05, 0.10, 0.20]
sens = my_sens(model, base_price, percents)

print("\n=== АНАЛІЗ ЧУТЛИВОСТІ ===")
display(sens)

plt.figure(figsize=(6,4))
plt.plot(sens["Δ_input_%"], sens["Δ_output_%"], marker="o")
plt.xlabel("Зміна ціни на ринку A (%)")
plt.ylabel("Зміна прогнозу ринку B (%)")
plt.title("Аналіз чутливості")
plt.grid()
plt.show()
```

**Завдання 8. Аналіз впливу валютних курсів**

 Виконано перерахунок цін із локальної валюти XOF у долари США за фіксованим курсом 1 XOF = 0.0016 USD. На основі середніх річних значень побудовано дві лінійні регресійні моделі: в локальній валюті та в USD.

Графіки показують, що форма тренду повністю зберігається після конвертації у USD, а всі значення масштабу просто зменшуються пропорційно курсу. Це підтверджується коефіцієнтом кореляції 1.0 між локальними цінами та цінами в USD, що означає ідеальну лінійну відповідність.

Метрики моделей також ідентичні за якістю прогнозу: значення R² = 0.52 однакові в обох варіантах, що вказує на те, що зміна валюти не впливає на пояснювальну здатність моделі. Відмінності спостерігаються лише у масштабі помилки: наприклад, RMSE у USD істотно менше через перерахунок у менші одиниці виміру.

Коефіцієнт нахилу моделі в локальній валюті становить 7.72, тоді як в USD — 0.012, що повністю відповідає застосованому курсу конвертації.

Отже, переведення цін у USD не змінює характер тренду, форму кривої та якість прогнозування, а потрібне виключно для зручності інтерпретації та міжнародного порівняння.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

CURRENCY = "XOF"
fx = {
    "AFN": 0.011,
    "INR": 0.012,
    "XOF": 0.0016,
    "PHP": 0.018,
    "MXN": 0.059,
    "COP": 0.00026,
    "ETB": 0.018,
    "ZAR": 0.054,
    "THB": 0.028,
    "USD": 1.0
}

df_filt = df[(df["mp_year"] <= 2000) & (df["cur_name"] == CURRENCY)].copy()
print("Кількість записів:", len(df_filt))

df_filt["fx_to_usd"] = df_filt["cur_name"].map(fx)
print("\nКурс для валюти:", fx.get(CURRENCY))

df_filt["price_usd"] = df_filt["mp_price"] * df_filt["fx_to_usd"]

yearly = (
    df_filt
    .groupby("mp_year", as_index=False)[["mp_price","price_usd"]]
    .mean()
)

print("\n=== ДАНІ ПО РОКАХ ===")
display(yearly.head())

X = yearly[["mp_year"]].values
y_orig = yearly["mp_price"].values
y_usd = yearly["price_usd"].values

lr_orig = LinearRegression().fit(X, y_orig)
lr_usd = LinearRegression().fit(X, y_usd)

pred_orig = lr_orig.predict(X)
pred_usd = lr_usd.predict(X)


def metrics(y, yhat):
    return {
        "MAE": mean_absolute_error(y, yhat),
        "RMSE": np.sqrt(mean_squared_error(y, yhat)),
        "R²": r2_score(y, yhat)
    }

print("\n=== МЕТРИКИ ===")
print("Original:", metrics(y_orig, pred_orig))
print("USD:", metrics(y_usd, pred_usd))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(yearly["mp_year"], y_orig, "o-")
plt.plot(yearly["mp_year"], pred_orig, "--")
plt.title("Ціни у локальній валюті")
plt.xlabel("Рік")
plt.ylabel("mp_price")
plt.grid()

plt.subplot(1,2,2)
plt.plot(yearly["mp_year"], y_usd, "o-")
plt.plot(yearly["mp_year"], pred_usd, "--")
plt.title("Ціни в USD")
plt.xlabel("Рік")
plt.ylabel("USD price")
plt.grid()

plt.tight_layout()
plt.show()

print("\n=== КОЕФІЦІЄНТИ ===")
print("Original slope:", lr_orig.coef_[0])
print("USD slope:", lr_usd.coef_[0])
print("Кореляція:", np.corrcoef(yearly["mp_price"], yearly["price_usd"])[0,1])

```

**Завдання 10. Порівняльний аналіз груп товарів**

Для порівняльного аналізу товари було згруповано у дві категорії: Grains (зернові культури) та Vegetables (овочева група). Аналіз динаміки показав, що ціни категорії Grains характеризуються різким зростанням у 1996–1998 роках із подальшим спадом у 1999–2000 рр. Водночас група Vegetables демонструє більш стабільну зміну цін, без різких піків. Це свідчить про більшу волатильність зернових ринків порівняно з овочами, що може пояснюватися залежністю зернових від кліматичних факторів, урожайності та імпорту.

Для оцінки взаємозв’язку між групами побудовано лінійну регресійну модель:

                 Vegetables = 0.066 · Grains + 146.64

Коефіцієнт детермінації R² = 0.035 свідчить про дуже слабку залежність між цінами двох товарних груп. Отримане значення нахилу регресії підтверджує, що зміна цін у групі Grains майже не впливає на динаміку Vegetables.

Таким чином, ринки зернових та овочів у досліджуваний період поводяться переважно незалежно, а цінові коливання в одній категорії не можна ефективно використовувати для прогнозування іншої. Це підтверджується як низьким значенням R², так і характером розсіювання точок на scatter-плоті.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

GROUP_1 = "Grains"
GROUP_2 = "Vegetables"

COUNTRY = None

grains_kw = [
    "rice", "maize", "wheat", "millet", "sorghum", "teff",
    "wheat flour", "maize flour"
]

vegetables_kw = [
    "beans", "onions", "potatoes", "cassava", "chickpeas"
]

df = pd.read_csv("global_food_prices_usd.csv", low_memory=False)
df = df[df["mp_year"] <= 2000].copy()

if COUNTRY:
    df = df[df["adm0_name"] == COUNTRY]

print("Кількість записів:", len(df))

def extract_group(df, keywords, name):
    mask = df["cm_name"].str.lower().str.contains("|".join(keywords), na=False)
    g = df[mask].copy()

    print(f"{name}: знайдено {len(g)} записів")

    yearly = (
        g.groupby("mp_year", as_index=False)["mp_price"]
        .mean()
        .rename(columns={"mp_price": name})
    )
    return yearly

grains = extract_group(df, grains_kw, GROUP_1)
vegetables = extract_group(df, vegetables_kw, GROUP_2)

data = pd.merge(grains, vegetables, on="mp_year")

print("\n=== ПЕРШІ РЯДКИ ===")
print(data.head())

print("\n=== ОПИСОВА СТАТИСТИКА ===")
print(data.describe())

X = data[[GROUP_1]].values
y = data[GROUP_2].values

model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)

print("\n=== РІВНЯННЯ РЕГРЕСІЇ ===")
print(f"{GROUP_2} = {model.coef_[0]:.3f} * {GROUP_1} + {model.intercept_:.2f}")
print("R²:", model.score(X, y))

plt.figure(figsize=(10,5))
plt.plot(data["mp_year"], data[GROUP_1], marker="o", label=GROUP_1)
plt.plot(data["mp_year"], data[GROUP_2], marker="o", label=GROUP_2)
plt.title("Тренди цін двох товарних груп")
plt.xlabel("Рік")
plt.ylabel("Середня ціна")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(6,5))
plt.scatter(X, y)
plt.plot(X, pred, color="red")
plt.title("Регресійна залежність між групами")
plt.xlabel(GROUP_1)
plt.ylabel(GROUP_2)
plt.grid()
plt.show()
```
